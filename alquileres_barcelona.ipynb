```python
df_1 = pd.read_csv("ficheros/2014_lloguer_preu_trim.csv")
df_2 = pd.read_csv("ficheros/2015_lloguer_preu_trim.csv")
df_3 = pd.read_csv("ficheros/2016_lloguer_preu_trim.csv")
df_4 = pd.read_csv("ficheros/2017_lloguer_preu_trim.csv")
df_5 = pd.read_csv("ficheros/2018_lloguer_preu_trim.csv")
df_6 = pd.read_csv("ficheros/2019_lloguer_preu_trim.csv")
df_7= pd.read_csv("ficheros/2020_lloguer_preu_trim.csv")
df_8= pd.read_csv("ficheros/2021_lloguer_preu_trim.csv")
df_9 = pd.read_csv("ficheros/2022_lloguer_preu_trim.csv")
df_10 = pd.read_csv("ficheros/2023_lloguer_preu_trim.csv")
```
